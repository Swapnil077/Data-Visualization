# Categorical

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head()
```

# Bar Graph

```python
sns.countplot(data = df , x= "neighbourhood_group")
```

```python
sns.countplot(data = df , x= "room_type")
```

```python
sns.countplot(data = df , x= "neighbourhood_group", hue= "room_type")  # Number of room type is paticular city
```

# Pie Chart

```python
df["room_type"].value_counts().plot(kind="pie",autopct ="%.2f")
```

```python
df["neighbourhood_group"].value_counts().plot(kind="pie",autopct ="%.2f")
```

```python
sns.countplot(data = df, x = "room_type")
```
