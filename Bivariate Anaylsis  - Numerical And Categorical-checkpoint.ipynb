```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.barplot(x ="class",y= "age",data = df)
```

```python
sns.barplot(x ="class", y= "age",data = df, estimator = "sum")
```

```python
sns.barplot(x ="class",y= "fare", hue ="sex" ,data = df, estimator = "sum")
```

```python
sns.barplot(x ="class",y= "age", hue ="sex" ,data = df)
```

```python
sns.boxplot(x ="class",y= "age", hue ="sex" ,data = df)
```

```python
sns.boxplot(x ="class",y= "age", hue ="survived" ,data = df)
```

```python
sns.violinplot(x ="class",y= "age", hue ="survived" ,data = df)
```

```python
sns.distplot(df[df["class"] =="Second"]["age"])
```
