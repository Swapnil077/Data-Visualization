```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

# Scatterplot

```python
sns.scatterplot(x ="age",y ="fare",data =df)
```

```python
sns.scatterplot(x ="age",y ="fare",hue ="survived",data =df)
```

```python
df2 =sns.load_dataset("tips")
df2.head()
```

```python
sns.scatterplot(x ="total_bill",y ="tip",data =df2)
```

```python
sns.scatterplot(x ="total_bill",y ="tip",hue ="sex",data =df2)
```

```python
sns.scatterplot(x ="total_bill",y ="tip",hue ="sex",style ="smoker",data =df2)
```

# Jointplot

```python
sns.jointplot(x ="age",y ="fare",data =df)
```

```python
sns.jointplot(x ="total_bill",y ="tip",hue ="sex",data =df2)
```

```python
sns.jointplot(x ="total_bill",y ="tip",data =df2,kind ="reg")
```
